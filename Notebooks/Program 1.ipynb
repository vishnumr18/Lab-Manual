## Develop a program to create histograms for all numerical features and analyze the distribution of each feature. Generate box plots for all numerical features and Identify any outliers. Use California Housing dataset.

```python
# Importing libraries 
import pandas as pd
from sklearn.datasets import fetch_california_housing
import numpy as np
import matplotlib.pyplot as plt
```

### Importing the California Housing data

```python
# creating an object of california housing data Class
data = fetch_california_housing()
```

```python
data
```

```python
# Creating a Pandas DataFrame for the CHD (California Housing dataset)
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
# Setting the target column in the Dataframe
df['Target'] = data.target
```

```python
# Printing the DataFrame
df
```

### Plotting Histograms for all the Numerical datatypes

```python
# Number of features
num_features = len(df.columns)
num_features
```

```python
# Creating the subplots
fig, ax = plt.subplots(nrows=(num_features+1)//2, ncols=2, figsize=(14, 28))
ax = ax.flatten()

# creating histograms for all the numerical features
for i, feature in enumerate(df.columns):
    ax[i].hist(df[feature], bins=30, alpha=0.7, color='skyblue', edgecolor='black')

    mean_val = df[feature].mean()
    median_val = df[feature].median()
    ax[i].axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}')
    ax[i].axvline(median_val, color='green', linestyle='-.', label=f'Median: {median_val:.2f}')
    ax[i].legend()
    
    ax[i].set_title(f'{feature} Distribution')
    ax[i].set_xlabel(feature)
    ax[i].set_ylabel("Count")

for j in range(i + 1, len(ax)):
    fig.delaxes(ax[j])


fig.suptitle("Histograms for California Housing Features", fontsize=24, y=0.92)
```

### Plotting Box plot for all the Numerical datatypes

```python
# Creating the subplots
fig, ax = plt.subplots(nrows=(num_features+1)//2, ncols=2, figsize=(14, 28))
ax = ax.flatten()

## creating histograms for all the numerical feature
for i, feature in enumerate(df.columns):
        boxplot = ax[i].boxplot(df[feature], vert=False, patch_artist=True)

        ax[i].set_title(f'{feature} Box Plot')
        ax[i].set_xlabel("Count")
        ax[i].set_ylabel(feature)


for j in range(i+1, len(ax)):
        fig.delaxes(ax[j])

fig.suptitle("Box Plots for California Housing Features", fontsize=24, y=0.92);
```
